```python
# Using the elbow method to find the optimal number of clusters. Lower the WCSS, the better it is
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42, max_iter=300,n_init=10) ## n_init number of centroids initialized
    kmeans.fit(df.drop('Private',axis=1))
    wcss.append(kmeans.inertia_) ## kmeans.inertia_ = helps to compute WCSS

    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, init = 'k-means++')
y_kmeans = kmeans.fit_predict(df.drop('Private',axis=1))
print(y_kmeans)
```

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head() ## New column added called cluster
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],y_kmeans))
print(classification_report(df['Cluster'],y_kmeans))
```
