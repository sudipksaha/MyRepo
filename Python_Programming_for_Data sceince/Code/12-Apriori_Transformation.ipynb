```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.ExcelFile('Dataframe_Apriori_Transformation.xlsx')
df1 = df.parse('Sheet1')
df1
```

```python
df1=df1[['User Id','Subscribed Channel']]
```

```python

## Rename the columns
df1=df1.rename(index=str, columns={"Subscribed Channel": "SubscribedChannel"})
df1=df1.rename(index=str, columns={"User Id": "UserId"})

df1
```

```python
unique_Ch=df1.SubscribedChannel.unique() ## Check all unique channels
unique_Ch
```

```python
unique_usr_id=df1.UserId.unique() ## Check all unique users
unique_usr_id
```

```python
## Arrange the data
for usrId in range(len(unique_usr_id)): ##  for each unique user
    for var in range(len(unique_Ch)):  ## for each unique channel
        ## match user's subscribed channel with unique channel lists
        df1[unique_Ch[var]]=np.where(df1['SubscribedChannel']==unique_Ch[var], unique_Ch[var],0) 
        ## adding a new column and filling it up with the subscribed channel detail
```
